This is our notebook for our preliminary data cleaning and logistic regression model.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from dateutil import parser
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from datetime import datetime as DAMN
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

```

```python
df = pd.read_csv("crime.csv")
df.shape
```

```python
df.drop(['Unnamed: 0', 'Block', 'Beat', 'X Coordinate', 'Y Coordinate', 'Latitude', 'Longitude', 'Ward', 'Community Area'], inplace=True, axis=1)
```

```python
df[:100]
```

```python
# One Hot Encoded (Categorical Data)
df = pd.get_dummies(df, columns =['Primary Type', 'Description', 'Location Description', 'District'])
```

```python
df[:100]
```

```python
y = df["Arrest"].copy(True)
X = df.drop(['Arrest'], axis=1).copy(True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = DAMN.now().second)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
classifier = GaussianNB()
```

```python
y_pred = classifier.fit(X_train, y_train).predict(X_test)
confusion_matrix_result = confusion_matrix(y_test, y_pred)
print(confusion_matrix_result)
```

```python
classifier.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
classifier2 = SGDClassifier(loss='hinge', penalty='l1')
```

```python
y_pred2 = classifier2.fit(X_train, y_train).predict(X_test)
conf_mat2 = confusion_matrix(y_test, y_pred2)
print(conf_mat2)
```

```python
classifier2.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred2))
```

```python
classifier3 = DecisionTreeClassifier(random_state=DAMN.now().second)
```

```python
y_pred3 = classifier3.fit(X_train, y_train).predict(X_test)
conf_mat3 = confusion_matrix(y_test, y_pred3)
print(conf_mat3)
```

```python
classifier3.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred3))
```
