This is our notebook for our KNN model.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
from datetime import datetime as DAMN
```

```python
df = pd.read_csv("crime.csv")
print(df.size)
```

```python
df[:10]
```

```python
df.drop(["Unnamed: 0", "Block", "Beat", "Ward", "Community Area", "Latitude", "Longitude", "District"], inplace=True, axis=1)
```

```python
df[:5]
```

```python
# One Hot Encoded (Categorical Data): Primary Type, Description, Location Description
crime = pd.get_dummies(df, columns =['Primary Type', 'Description', 'Location Description'])
```

```python
# Look at data
crime[:10]
```

```python
# See columns present
len(crime.columns)
```

```python
y = crime["Arrest"].copy(True)
X = crime.iloc[:,1:].copy(True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state = DAMN.now().second)
```

```python
# instantiate a KNN classifier with 3 neighbors
knn = KNeighborsClassifier(n_neighbors=1000)
```

```python
# fit the model
knn.fit(X_train, y_train)
```

```python
# predict the response
pred = knn.predict(X_test)
```

```python
# get the accuracy score
accuracy_score(y_test, pred)
```

```python
# Now let's try fitting KNN for 1000, 2000, 3000, 4000, 5000 neighbors
neighbors = range(1000,6000,1000)
cv_scores = []
```

```python
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    cv_scores.append(accuracy_score(y_test, pred))
```

```python
# changing to misclassification error
MSE = [1 - x for x in cv_scores]

# determining best k
optimal_k = neighbors[MSE.index(min(MSE))]
print "The optimal number of neighbors is %d" % optimal_k

# plot misclassification error vs k
plt.plot(neighbors, MSE)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Misclassification Error')
plt.show()
```
